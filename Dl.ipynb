```python
# ==========================================================
# DEEP LEARNING–BASED DRIVER ACTIVITY RECOGNITION (VISIONDRIVE)
# Based on: Obaid et al., IEEE Access 2024
# ==========================================================
!pip install tensorflow opencv-python matplotlib seaborn scikit-learn --quiet

import numpy as np, matplotlib.pyplot as plt, seaborn as sns, tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (ConvLSTM2D, BatchNormalization, MaxPooling3D,
                                     Dropout, Flatten, Dense, Input, Multiply,
                                     TimeDistributed, Conv2D, MaxPooling2D, LSTM, Concatenate)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.metrics import (classification_report, confusion_matrix,
                             f1_score, precision_score, recall_score, roc_curve, auc)
from sklearn.preprocessing import label_binarize

# -----------------------------------------------------------
# CONFIGURATION
# -----------------------------------------------------------
IMG_SIZE, SEQ_LEN, NUM_CLASSES = 64, 15, 10
SAMPLES = 80   # small synthetic demo dataset (to avoid Colab crash)
EPOCHS = 10
BATCH = 4
np.random.seed(42)

# -----------------------------------------------------------
# SYNTHETIC DATA (simulating VisionDrive dataset)
# -----------------------------------------------------------
X = np.random.rand(SAMPLES, SEQ_LEN, IMG_SIZE, IMG_SIZE, 3).astype("float16")
y = np.random.randint(0, NUM_CLASSES, SAMPLES)
gender = np.random.randint(0, 2, SAMPLES)  # 0=male, 1=female
y = to_categorical(y, NUM_CLASSES)
gender = np.expand_dims(gender, 1)

# Split
X_train, X_test, y_train, y_test, g_train, g_test = train_test_split(
    X, y, gender, test_size=0.25, random_state=42
)

# -----------------------------------------------------------
# ATTENTION BLOCK
# -----------------------------------------------------------
def attention_block(inputs):
    att = Dense(inputs.shape[-1], activation='softmax')(inputs)
    return Multiply()([inputs, att])

# -----------------------------------------------------------
# ConvLSTM MODEL (with diversity-aware fusion)
# -----------------------------------------------------------
def build_conv_lstm(n_classes):
    vid_in = Input(shape=(SEQ_LEN, IMG_SIZE, IMG_SIZE, 3))
    g_in   = Input(shape=(1,))  # gender metadata

    x = ConvLSTM2D(8, (3,3), activation='tanh', return_sequences=True)(vid_in)
    x = BatchNormalization()(x)
    x = MaxPooling3D((1,2,2))(x)
    x = Dropout(0.25)(x)

    x = ConvLSTM2D(16, (3,3), activation='tanh')(x)
    x = Flatten()(x)
    x = attention_block(x)
    g = Dense(8, activation='relu')(g_in)
    x = Concatenate()([x, g])

    x = Dense(128, activation='relu')(x)
    x = Dropout(0.3)(x)
    out = Dense(n_classes, activation='softmax')(x)

    model = Model([vid_in, g_in], out)
    model.compile(optimizer=Adam(1e-4), loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# -----------------------------------------------------------
# LRCN MODEL (CNN + LSTM late fusion)
# -----------------------------------------------------------
def build_lrcn(n_classes):
    vid_in = Input(shape=(SEQ_LEN, IMG_SIZE, IMG_SIZE, 3))
    g_in   = Input(shape=(1,))

    x = TimeDistributed(Conv2D(16, (3,3), activation='relu'))(vid_in)
    x = TimeDistributed(MaxPooling2D(2,2))(x)
    x = TimeDistributed(Conv2D(32, (3,3), activation='relu'))(x)
    x = TimeDistributed(MaxPooling2D(2,2))(x)
    x = TimeDistributed(Flatten())(x)
    x = LSTM(128)(x)
    x = attention_block(x)

    g = Dense(8, activation='relu')(g_in)
    x = Concatenate()([x, g])
    x = Dense(128, activation='relu')(x)
    x = Dropout(0.3)(x)
    out = Dense(n_classes, activation='softmax')(x)

    model = Model([vid_in, g_in], out)
    model.compile(optimizer=Adam(1e-4), loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# -----------------------------------------------------------
# SELECT MODEL (Change here: ConvLSTM or LRCN)
# -----------------------------------------------------------
USE_LRCN = False
model = build_lrcn(NUM_CLASSES) if USE_LRCN else build_conv_lstm(NUM_CLASSES)
model.summary()

# -----------------------------------------------------------
# TRAINING
# -----------------------------------------------------------
history = model.fit([X_train, g_train], y_train,
                    validation_split=0.2,
                    epochs=EPOCHS, batch_size=BATCH, verbose=1)

# -----------------------------------------------------------
# EVALUATION
# -----------------------------------------------------------
y_prob = model.predict([X_test, g_test])
y_pred = np.argmax(y_prob, axis=1)
y_true = np.argmax(y_test, axis=1)

# Metrics
print("\nCLASSIFICATION REPORT\n")
print(classification_report(y_true, y_pred, digits=3))
print("Weighted Precision:", precision_score(y_true, y_pred, average='weighted'))
print("Weighted Recall:", recall_score(y_true, y_pred, average='weighted'))
print("Weighted F1:", f1_score(y_true, y_pred, average='weighted'))

# -----------------------------------------------------------
# GRAPHS SECTION
# -----------------------------------------------------------

# 1️⃣ Accuracy and Loss Curves
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], '-o', label='Training Accuracy')
plt.plot(history.history['val_accuracy'], '-o', label='Validation Accuracy')
plt.xlabel('Epochs'); plt.ylabel('Accuracy'); plt.legend()
plt.title('Training vs Validation Accuracy')

plt.subplot(1,2,2)
plt.plot(history.history['loss'], '-o', label='Training Loss')
plt.plot(history.history['val_loss'], '-o', label='Validation Loss')
plt.xlabel('Epochs'); plt.ylabel('Loss'); plt.legend()
plt.title('Training vs Validation Loss')
plt.show()

# 2️⃣ Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(7,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='YlGnBu',
            xticklabels=[f'Class {i}' for i in range(NUM_CLASSES)],
            yticklabels=[f'Class {i}' for i in range(NUM_CLASSES)])
plt.xlabel('Predicted Labels'); plt.ylabel('True Labels')
plt.title('Confusion Matrix - Driver Activity Recognition')
plt.show()

# 3️⃣ Per-Class Precision, Recall, F1
report = classification_report(y_true, y_pred, output_dict=True)
classes = list(report.keys())[:-3]
precision = [report[c]['precision'] for c in classes]
recall = [report[c]['recall'] for c in classes]
f1 = [report[c]['f1-score'] for c in classes]

x = np.arange(len(classes))
plt.figure(figsize=(10,4))
plt.bar(x - 0.25, precision, width=0.25, label='Precision')
plt.bar(x, recall, width=0.25, label='Recall')
plt.bar(x + 0.25, f1, width=0.25, label='F1-score')
plt.xticks(x, classes, rotation=45)
plt.title('Per-Class Metrics')
plt.legend()
plt.tight_layout()
plt.show()

# 4️⃣ ROC Curves for each class (multi-class)
y_true_bin = label_binarize(y_true, classes=np.arange(NUM_CLASSES))
fpr, tpr, roc_auc = {}, {}, {}
plt.figure(figsize=(8,6))
for i in range(NUM_CLASSES):
    fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_prob[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
    plt.plot(fpr[i], tpr[i], lw=1.5, label=f'Class {i} (AUC={roc_auc[i]:.2f})')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate'); plt.ylabel('True Positive Rate')
plt.title('ROC Curves per Class')
plt.legend()
plt.show()

# 5️⃣ Class Distribution Visualization
plt.figure(figsize=(7,4))
sns.countplot(x=np.argmax(y, axis=1))
plt.title("Synthetic Dataset: Class Distribution (VisionDrive Simulation)")
plt.xlabel("Driver Activity Class"); plt.ylabel("Samples per Class")
plt.show()

print("\n✅ Training complete — model & analytics generated successfully!")
```
